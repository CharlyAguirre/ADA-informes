# 1. Problema corte de varilla.

**Entrada: listaPrecio**: arreglo con los precios de los cortes que se le pueden realizar a la varilla, **largoVar** : largo de la varilla de la cual buscamos sacarle el mayor beneficio haciéndole o no cortes. 

**def cortarVarilla(listaPrecio, largoVar)**

**Salida: tabla[largoVar]**: en esta variable se guarda el máximo beneficio posible de la varilla de largoVar, **cortes**: esta variable contiene la cantidad de cortes que se deben realizar a la varilla para llegar al maximo beneficio posible

**return tabla[largoVar], cortes**

El problema de corte de varilla consiste en que nos entregan una varilla de largo n a la cual se le pueden realizar cortes específicos y tienes los precios a los que se vender las distintas varillas que se obtienen al realizarle cortes a la varilla, también esta la posibilidad de no realizarle ningún corte a la varilla. Este es un problema de optimización, pero ¿en qué consiste un problema de optimización? Estos problemas buscan en maximizar o minimizar una variable, en este caso se busca maximizar el beneficio por lo que calza con la definición de este problema.

# 2. corteVarilla

El algoritmo para obtener la solución al problema de corte de varilla tiene varias formas de ser implementado, las cuales son:

1. **Iterativo bottom-up** (programación dinámica)
2. Recursivo top-down (programación dinámica)
3. **Recursivo**

En este informe se implementarán tanto el método recesivo y el método iterativo bottom-up.

**1)** Un algoritmo iterativo bottom-up es un método de programación dinámica. La programación dinámica consiste en la combinación de soluciones de sub-promeblas para encontrar la solución al problema mayor al que pertenecen estos sub-problemas, las soluciones de estos sub-problemas son guardadas por si más adelante el algoritmo se encuentra con un sub-problema igual a uno ya calculado no tenga que calcularlo otra vez y solo acceda a la solución, estas soluciones son guardadas en un arreglo, diccionario, tabla hash, etc. Por lo mencionado anteriormente es recomendable usar este método cuando al dividir el problema mayor en sub-problemas se observe que en el desarrollo del algoritmo este se encontrara con sub-problemas iguales, ya que si todos los sub-problemas son distintos sería un gasto de memoria innecesario el realizar esta implementación.

Que la programación dinámica sea bottom-up se refiere a que los sub-problemas se van resolviendo desde el más pequeño al más grande.

## 2.1 descripción corteVarilla bottom-up.

Como fue explicado en el punto anterior usaremos la forma de implementación iterativamente bottom-up.

Primero empieza este algoritmo por declarar un arreglo en que se almacenaran el máximo beneficio posible para una varilla que tenga como largo el índice de este **(ej: arreglo[1] = 1, una varilla de largo 1 tiene como máximo beneficio 1)**, este arreglo será inicializado todo lleno de ceros. **(arreglo de cortes)** 

Luego el algoritmo sigue por un ciclo repetitivo for empezando desde **1** hasta **largoVar + 1**, este ciclo repetitivo representa la varilla actual a la cual se le va a calcular su máximo beneficio posible **(ej: varillaActual = 3 se le calcula el máximo beneficio a una varilla de largo 3)**. Para calcular el máximo beneficio de la varilla actual usamos otro ciclo repetitivo for empezando desde **1** hasta **varillaActual + 1**, la i que se mueve por este ciclo representa a que altura se le realizara el corte a la varilla actual. Este ciclo para calcular el mejor valor busca el valor de una varilla de largo **i** en la lista de precios y le suma el valor del máximo beneficio de la varilla restante **(ej: varrillaActual = 3, i = 2, valor = listaPrecios[2] + arreglo[3-2=1])*. Estos dos ciclos empiezan por el 1 ya que es un algoritmo bottom-up, empiezan calculando el sub-problema más pequeño y con este dato van construyendo las soluciones siguientes.

El algoritmo repite el proceso de calcular el máximo beneficio de la varilla actual hasta llegar a que varillaActual sea igual al largo ingresado.

## 2.3 Codigo corteVarilla bottom-up

```python
def cortarVarilla(largoVar, listaPrecio, verbose = False):

  arreglo = [0] * (largoVar + 1) #arreglo donde se guardan los maximos beneficios
  cortes = [0] * (largoVar + 1) # arreglo donde se guardan los cortes necesarios para optener los maximos beneficio
  problema = 0

  for varillaActual in range(1, largoVar + 1): # for para saber cual sub-problema estamos resolviendo 

    if(verbose == True):
      print("La varilla a la que se calculara su maximo beneficio es:", varillaActual)
      print("Empezamos con el mejor valor =", arreglo[varillaActual])


    for i in range(1, varillaActual+1): # for para sacar el maximo beneficio posible a la varillaActual

      problema = problema + 1 #problemas a resolver

      valor = listaPrecio[i] + arreglo[varillaActual-i]

      if(verbose == True):
        print("Si cortamos la varilla actual:", varillaActual, "en el punto ", i)
        print("Nos quedan 2 varilla, una de largo", i , "y otra de largo ", (varillaActual - i))
        print("valor de estas dos varillas es", listaPrecio[i] ,"+",  arreglo[varillaActual-i], "=", valor)
        print("valor > mejorValor?", valor, ">", arreglo[varillaActual])
        print(" ")

      if (valor > arreglo[varillaActual]) :
        arreglo[varillaActual] = valor

        if(verbose == True):
          print("SI")
          print("Ahora actrualizamos el mejor valor =", valor)
          print("Preguntamos i < largoVar para saber si al mejor valor se le realizo un corte")
          print(i,"<",  largoVar)
          print(" ")

        if (i < largoVar): # si se cumple esta condiciion significa que se realizo un corte
          cortes[varillaActual] = 1 + cortes[varillaActual - i]

          if(verbose == True):
            print("SI")
            print("Actualizamos el arreglo de cortes en la posicion de la varilla actual")
            print("cortes[varillaActual] =", 1, "+", cortes[varillaActual - 1], "=", cortes[varillaActual])
            print(" ")
        else: # como no  se cumple esta condicion anterio significa que no se realizaron cortes
          cortes[varillaActual] = 0

          if(verbose == True):
            print("NO")
            print("No se realizaron cortes")
            print(" ")

    if(verbose == True):
      print("Al terminar el for para calcular el maximo beneficio actualizamos en el arreglo donde se guardan estos datos")
      print("arreglo =", arreglo)
      print(" ")


  return arreglo[largoVar], cortes[largoVar], problema



lista_p = [0,1,5,8,9,10,17,17,20,24,30]

n = 6

maximo, cortes, problema = cortarVarilla(n, lista_p)

print("Largo de la varilla =", n)
print("Lista de precios =", lista_p)
print("Maximo beneficio para la varilla =", maximo)
print("Cortes realizados =", cortes)
print("Problemas resueltos =", problema)
```

## 2.6 Codigo corteVarilla recursivo.

```python
def corteVarillaRecur(largoVar, listaPrecio, cont = 0, corte = 0, verbose = False):
  
  #caso base
  if(largoVar == 0):
     return 0, cont, corte

  mejorVal = -9999999

  for i in range(1, largoVar + 1):

    if(verbose == True):
      print("La varilla a la que se calculara su maximo beneficio es:", i)
      print("Empezamos con el mejor valor =", mejorVal)

    valorAux, cont, cor= corteVarillaRecur(largoVar-i, listaPrecio , cont+1, corte, verbose)

    if(verbose == True):
        print("Si cortamos la varilla actual:", largoVar, "en el punto ", i)
        print("Nos quedan 2 varilla, una de largo", i , "y otra de largo ", (largoVar - i))
        print("valor de estas dos varillas es", listaPrecio[i] ,"+",  valorAux, "=", listaPrecio[i] + valorAux)
        print("valor > mejorValor?", listaPrecio[i] + valorAux, ">", mejorVal)
        print(" ")

    if(mejorVal < listaPrecio[i] + valorAux):

      mejorVal = listaPrecio[i] + valorAux

      if(verbose == True):
          print("SI")
          print("Ahora actrualizamos el mejor valor =", mejorVal)
          print("Preguntamos i < largoVar para saber si al mejor valor se le realizo un corte")
          print(i,"<",  largoVar)
          print(" ")

      if(i < largoVar):
        corte = cor + 1

        if(verbose == True):
            print("SI")
            print("Actualizamos el arreglo de cortes en la posicion de la varilla actual")
            print("cortes[varillaActual] =", 1 ,"+", cor ,"=", corte)
            print(" ")

      else:
        corte = 0

        if(verbose == True):
            print("NO")
            print("No se realizaron cortes")
            print(" ")

  if(verbose == True):
    print("Se retorna el mejor valor =", mejorVal)
    print("Se retorna la cantidad de llamas recursivas =", cont)
    print("Se retorna la cantidad de cortes necesarios =", corte)


  return mejorVal, cont, corte

lista_p = [0,1,5,8,9,10,17,17,20,24,30]

n = 5

cont = 0

maximo, cont, corte = corteVarillaRecur(n, lista_p, cont)

print("Largo de la varilla =", n)
print("Lista de precios =", lista_p)
print("Maximo beneficio para la varilla =", maximo)
print("La cantidad de llamadas recursivas es",  cont)
print("Cortes realizados =", corte)
```

# 3. Tiempo de ejecución. 

## 3.1 Teorema para el tiempo de ejecución para el algoritmo bottom-up.

El tiempo de ejecución de manera teórica tiene el algoritmo para encontrar el máximo beneficio posible de una varilla de largo n en notación O es de **O(n^2)** para todos sus casos.

## 3.2 Comprobación del teorema.

La ecuación de recurrencia **T(n) = Σ**(desde j = 1 hasta n)**Σ**(desde i = 1 hasta j) **C**.

La primera **sumatoria representa el primer for** que se encuentra en el algoritmo y la segundo **sumatoria representa el segundo for** que se encuentra en el algoritmo, la letra **C representa el coste de cada iteración** del segundo for.

Por propiedad de sumatorias la ecuación queda: **T(n) = C * Σ**(desde j = 1 hasta n)**j**.

Podemos usar la identidad de sumatoria: **Σ**(desde i = 1 hasta n) **i** = **(n(n+1)) / 2**

Aplicando la identidad nos queda: **T(n) = C * ((n (n+1)) / 2)**

Distribuimos la n y queda: **T(n) = C * ((n^2 + n) / 2)**

Ahora aplicamos la notación O grande, con esta podemos ignorar las contantes y quedarnos con la n de mayor grado por lo que nos queda la ecuación de recurrencia: T(n) = O(n^2)

Con esto podemos concluir que el teorema que fue propuesto en un inicio es correcto.


# 4. Complejidad espacial.

La complejidad espacial del algoritmo bottom-up es de O(n^2), pero ¿cómo llegamos a esto? 

Para esto debemos saber que la complejidad espacial es la cantidad de memoria que necesita al algoritmo para resolver el problema que se le es planteado, solo se contabiliza la memoria que usa que no es destinada al input del problema, en nuestro caso serían los parámetros que le son entregados a la función.

Ya teniendo claro que es la complejidad espacial analicemos el algoritmo. 

Si observamos el código del algoritmo podemos ver que cuando se llama a la función corteVarilla lo primero que este realiza es declarar dos arreglos auxiliares, en los puntos anteriores ya fue explicando para que se utilizan estos arreglos, en este punto nos interesa que estos arreglos son de largo n siendo n el largo de la varilla a la cual se le quiere calcular el máximo beneficio, por lo que la complejidad espacial es O(n^2) ya que se usan dos listas/arreglos de largo n adicional.


# 5. Correctitud del corteVarilla bottom-up.

## 5.2 Comprobación del teorema.

**Caso base:**

Cuando n = 0, ya se encuentra el mayor beneficio posible del subproblema ya que este valor siempre es 0, junto con los cortes realizados que también son 0.

**Paso inductivo:**

Se asume que los sub-problemas anteriores a la iteración n.

Tenemos **valor = listaPrecio[n] + arreglo[n-i]** y lo comparamos con **arreglo[n]** y de aquí tenemos dos opciones.

1)Cuando valor es mayor, el sub-problema **arreglo[n-i]** ya fue resuelto en anteriores iteraciones, se guarda ese valor en **arreglo[n]**, cuando se entra en esta condición se comparan **n** y **i** se tienen 2 opciones

1.1) Si i es menor que n se realiza lo siguiente cortes[n] = 1 + cortes[n-i] siendo cortes[n-i] la cantidad de cortes necesarios para obtener el máximo beneficio perteneciente a arreglo[n-i], se retorna arreglo[n] y cortes[n]

1.2) Si i es igual que n se realiza lo siguiente cortes[n] = 0 porque si i == n no es necesario realizar cortes para obtener el máximo beneficio perteneciente a arreglo[n-i], se retorna arreglo[n] y cortes[n]

2) Cuando **arreglo[n]** es mayor, retorna el máximo beneficio del precio correspondiente del largo de la varilla completo y se retorna la cantidad de cortes necesario para ese valor que se guarda en **cortes[n]**, por lo que es correcto.
