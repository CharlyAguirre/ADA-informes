**def cortarVarilla(listaPrecio, largoVar)**

Este es un problema de optimizacion, pero en que consite un problema de optimizacion? Estos problemas buscan en maximizar o minimizar una variable, en este caso se busca maximizar el beneficion por lo que calza con la definicion de este problema.

# 2. corteVarilla

El algoritmo para optener la solucion al problema de corte de varilla tiene varia formas de ser implementado, las cuales son:

1. **Recusivo**
2. Recusivo top-down (programacion dinamica)
3. **Iterativo bottom-up** (programacion dinamica)

En este infome se implementaran tanto el metodo recusivo y el metodo iterativo bottom-up.

Un algoritmo iterativo bottom-up es un metodo de programacion dinamica. La programacion dinamica consiste de en la combinacion de soluciones de sub-promeblas para encontrar la solucion al problema mayor al que pretenecen estos sub-problemas, las soluciones de estos sub-problemas son guardadas por si mas adelante el algoritmo se encuentra con un sub-problema igual a uno ya calculado no tenga que calcularlo otra vez y solo acceda a la solucion, estas soluciones son guardaas en un arreglo, diccionario, tabla hash, etc. Por lo mencionado anteriormente es recomendable usar este metodo cuando al dividir el problema mayor en sub-problemas se observe que en el desarrollo del algorimo este se encontrara con sub-problemas iguales, ya que si todos los sub-problemas son distintos seria un gasto de memoria inneseriao el realizar esta implementacion.

Que la progrmacion dinamica sea bottom-up se refiere a que los sub-problemas se van resolviendo desde el mas pequenyo al mas grande.

## 2.1 Descripcion corteVarilla bottom-up.

Como fue explicado en el punto anterior usaremos la forma de implementacin iterativamente bottom-up.

Primero empieza este algortimo por declara un arreglo en que que se almacenaran el maximo beneficio posible para una varilla que tenga como largo el indice de este **(ej: arreglo[1] = 1, una varilla de largo 1 tiene como maximo beneficio 1)**, este arreglo sera inicializado todo lleno de ceros. **(arreglo de cortes)** 

Luego el algoritmo sigue por un ciclo repetitivo for empezando desde **1** hasta **largoVar + 1**, este ciclo repetitivo representa la varilla actual a la cual se le va a calcular su maximo beneficio posible **(ej: varillaActual = 3 se le calcula el maximo beneficio a una varilla de largo 3)**. Para calcular el maximo beneficio de la varilla actual usamos otro ciclo repetitivo for empezando desde **1** hasta **varillaActual + 1**, la i que se mueve por este ciclo representa a que altura se le realizara el corte a la varilla actual. Este ciclo para calcular el mejor valor busca el valor de una varilla de largo **i** en la lista de precios y le suma el valor del maximo beneficio de la varilla restante **(ej: varrillaActual = 3, i = 2, valor = listaPrecios[2] + arreglo[3-2=1])*. Estos dos ciclos empiezan por el 1 ya que es un algoritmo bottom-up, empiezan calculando el sub-problema mas pequenyo y con este dato van construllendo las soluciones siguientes.

El arlgoritmo repite el proceso de calcular el maximo beneficio de la varilla actual hasta llegar a que varillaActual sea igual al largo ingresado.
